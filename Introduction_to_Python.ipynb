```python
#function with input and with return value
def compute_num(numbers): #def nameoffunction(input params):
  c=len(numbers)
  total=0
  for n in numbers:
    total+=n
  return total,c
  # print("hey ",total) #ucant store the value
#function call
compute_num({1,2})
```

```python
op=compute_num([12,2]) #since usage of return
```

```python
op
```

```python
def print_the_name(name,times=5):   #times is a default param even if defined or not in caller function it calls the default times
  for i in range(times):
    print(name)

print_the_name("kk",3)
```
